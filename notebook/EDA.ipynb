```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('insurance.csv')
```

#### Dataset description:
1. Age: Person's age in years
2. Sex: Gender of the person or insurance holder(Female or Male)
3. BMI: Body mass index. The ideal range according to height and weight is 18.5 to 24.9
4. Children: Number of dependents
5. Smoker: Whether the insurance holder smokes or not
6. Region: Residential area of the person
7. Expenses: Individual medical costs billed by health insurance

```python
fig, ax =plt.subplots(1,2)
sns.countplot(df['sex'], ax=ax[0],palette='flare')
sns.countplot(df['region'], ax=ax[1],palette='flare')
plt.xticks(rotation=30)
```

```python
fig, ax =plt.subplots(1,2)
sns.countplot(df['children'], ax=ax[0],palette='flare')
sns.countplot(df['smoker'], ax=ax[1],palette='flare')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (15,10))

df.plot(kind='hist', y='age', ax=axes[0][0])
df.plot(kind='hist', y='bmi', ax=axes[0][1], bins = 54)
df.plot(kind='hist', y='children', ax=axes[1][0], bins = 6)
df.plot(kind='hist', y='expenses', ax=axes[1][1], bins = 80)
```

```python
fig, axes = plt.subplots(ncols = 3, figsize = (15,6), squeeze=True)
sns.scatterplot(x='bmi', y='expenses', ax=axes[0], data=df,hue='sex', palette='flare')

sns.scatterplot(x='bmi', y='expenses', ax=axes[1], data=df,hue='smoker', palette='flare')

sns.scatterplot(x='bmi', y='expenses', ax=axes[2], data=df,hue='region',palette='flare')
```
